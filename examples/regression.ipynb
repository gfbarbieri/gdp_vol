## Load Data from Data Directory

Read transformed data.

```python
excel_loader = DataRetriever.create(
    path_or_url='../data/transformed_data.xlsx'
)

data = excel_loader.load_data()

print(data.shape)
data.head()
```

## Regressions

```python
# Initialize the regressor.
regressor = OLSRegressor(dataframe=data)
```

```python
cols = [
    ('real_gdp_hp_cycle', ['real_gdpp_hp_cycle']),
    ('real_gdp_bk_cycle', ['real_gdpp_bk_cycle']),
    ('real_gdp_log_hp_cycle', ['real_gdpp_log_hp_cycle']),
    ('real_gdp_log_bk_cycle', ['real_gdpp_log_bk_cycle'])
]

# Loop through the columns and fit the model.
for y_col, x_cols in cols:
    
    # Fit the model.
    print(f"Fitting {y_col} on {x_cols}.")
    regressor.fit(y_col=y_col, x_cols=x_cols)

    # Print the summary.
    print(regressor.results.summary())
```
